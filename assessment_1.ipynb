 Write a function apply_discount(orders, discount_rate) that takes
a list of orders and applies a percentage discount (e.g., 10% = 0.10) to
the price_per_item for each order.


```python

discount_rate = 0.1 #global variable
def apply_discount(orders):
    global discount_rate #using global variable inside function
    discounted_orders = []
    for order in orders:
        discounted_price = order["price_per_item"] * order["quantity"] * discount_rate
        discounted_price=order["price_per_item"] * order["quantity"] - discounted_price
        #discounted_price is local variable

        discounted_order = {
            "order_id": order["order_id"],
            "customer": order["customer"],
            "item": order["item"],
            "quantity": order["quantity"],
            "price_per_item": discounted_price,
            "order_date": order["order_date"]
          }
        discounted_orders.append(discounted_order)

    return discounted_orders
```

```python
discounted_orders = apply_discount(sales_data)
print(discounted_orders)
```

Write a custom iterator class OrderIterator to iterate through the
orders, yielding one order at a time.

```python
class OrderIterator:
  def __init__(self,orders):
      self.orders=orders

  def print_orders(self):
      for order in self.orders:
          print(order)
```
