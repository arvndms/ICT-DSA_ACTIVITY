Bank Account Manager

```python

balance = 0

def deposit(amount):
    global balance
    balance += amount
    print(f"Deposited: {amount}")
    print(f"New Balance: {balance}\n")


def withdraw(amount):
    global balance
    if amount > balance:
        print("Insufficient balance!\n")
    else:
        balance -= amount
        print(f"Withdrew: {amount}")
        print(f"New Balance:{balance}\n")


def check_balance():
    print(f"Current Balance:{balance}\n")



print("Welcome to Bank Account Manager")
while True:
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance ")
    print("4. Quit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        amt = float(input("Enter amount to deposit: "))
        deposit(amt)

    elif choice == 2:
        amt = float(input("Enter amount to withdraw: "))
        withdraw(amt)

    elif choice == 3:
        check_balance()

    elif choice == 4:
        break

    else:
        print("Invalid choice! \n")
```

Weather Data Analyzer

```python
temp=[27,24,22,30,28,20]
humidity=[50,45,55,70,65,40]

def avg_temp(temp):
  return sum(temp)/len(temp)

def avg_humidity(humidity):
  return sum(humidity)/len(humidity)

avg_humidity=avg_humidity(humidity)
avg_temp=avg_temp(temp)
print('Average Temperature:',avg_temp)
print('Average Humidity:',avg_humidity)
print('Weather Condition:')
if avg_humidity>60:
  print('Rainy')
elif avg_temp>25:
  print('Sunny')
else:
  print('Cold')
```
