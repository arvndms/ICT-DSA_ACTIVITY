Encoding Categorical Variables

Ordinal Encoding

```python
df['Condition of the House'].value_counts()
```

```python
house_condition = {
    'Fair': 3,
    'Good': 4,
    'Excellent': 5,
    'Okay' : 2,
    'Bad': 1
}
```

```python
df['Condition of the House'] = df['Condition of the House'].map(house_condition)
```

```python
df['No of Times Visited'].nunique()
```

```python
times_visited = {
    0	: 0 ,
'Twice':	2,
'Thrice':	3,
'Once':	1,
'Four': 4
}
```

```python
df['No of Times Visited'] = df['No of Times Visited'].map(times_visited)
```

Nominal encoding

```python
df = pd.get_dummies(df,columns=['Waterfront View'],dtype=int)
```

Outlier Removal

```python
cols_to_check = ['Sale Price', 'No of Bedrooms', 'No of Bathrooms', 'Flat Area (in Sqft)',
                 'Lot Area (in Sqft)', 'Basement Area (in Sqft)', 'Age of House (in Years)']
```

```python
df.shape
```

```python
for cols in cols_to_check:
  Q1 = df[cols].quantile(0.25)
  Q3 = df[cols].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5* IQR
  upper_bound = Q3 + 1.5 * IQR
  df = df[(df[cols] >= lower_bound) & (df[cols] <= upper_bound)]
```
